```python
import pickle
import numpy
```

```python
outputs = [
    pickle.load(open('../input/tmp-feature-vector-dataset/feature_vector.pkl', 'rb')),
    pickle.load(open('../input/tmp-feature-vector-dataset/feature_vector(1).pkl', 'rb')),
    pickle.load(open('../input/tmp-feature-vector-dataset/feature_vector(2).pkl', 'rb')),
    pickle.load(open('../input/tmp-feature-vector-dataset/feature_vector(3).pkl', 'rb')),
    pickle.load(open('../input/tmp-feature-vector-dataset/feature_vector(4).pkl', 'rb')),
    pickle.load(open('../input/tmp-feature-vector-dataset/feature_vector(5).pkl', 'rb')),
    pickle.load(open('../input/tmp-feature-vector-dataset/feature_vector(6).pkl', 'rb')),
]
real_output = {}
```

```python
for output in outputs:
    for a in output: # model name
        if a not in real_output:
            real_output[a] = {}
        for b in output[a]: # dataset
            if b not in real_output[a]:
                real_output[a][b] = {}
            for c in output[a][b]: # scorer_name
                real_output[a][b][c] = output[a][b][c]    
```

```python
cnt = 0
for a in real_output:
    for b in real_output[a]:
        for c in real_output[a][b]:
            print(a, b, c)
            cnt += 1
```
