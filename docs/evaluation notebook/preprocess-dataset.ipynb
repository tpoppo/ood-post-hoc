```python
import tensorflow as tf
import numpy as np
import pickle
```

```python
paths = [
    ('imagenet', '../input/ilsvrc2012-validation-set/Images/imagenet'),
    ('iNaturalist', './iNaturalist/images'),
    ('SUN', './SUN/images'),
    ('Places', './Places/images'),   
]
ans = {}
```

```python
for name, path in paths:
    print(name, path)
    ds = tf.keras.utils.image_dataset_from_directory(
        directory=path,
        labels=None,
        image_size=(224, 224),
        batch_size=512,
    )
    print(ds)
    x = np.concatenate([a.numpy() for a in ds.take(4)], axis=0).astype('uint8')
    print(x.shape)
    ans[name] = x
    
```

```python
pickle.dump(ans, open('dataset.pkl', 'wb'))
```
